Data Preparation and Train-Test Split Code for Dunn Humby Data.
This file was ran 4 times for quarterly split data. Thus we have 4 train and 4 test files.

```python
import pandas as pd
import numpy as np
import collections
import glob
```

```python
path = "" # Data folder path for each quarter which contains the csv files for all 13 weeks of that quarter

all_files = glob.glob(path + "/*.csv") # Reading all csv files in that folder

l = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df = df[['SHOP_WEEK','PROD_CODE','CUST_CODE','BASKET_ID']] # Taking the important features only
    l.append(df)
    print(l)

data = pd.concat(l, axis=0, ignore_index=True)
```

```python
data = data.dropna()
data.head()
```

```python
data = data.sort_values(["CUST_CODE", "SHOP_WEEK", "BASKET_ID"],ascending=True)
data = data.reset_index(drop = True)
```

```python
# Loop to generate order_number
bnum = []
cust = []
bas = []
n = 1
for i in range(len(data)):
    print(len(bnum))
    if data["CUST_CODE"][i] in cust:
        if data["BASKET_ID"][i] in bas:
            bnum.append(n)
        else:
            bas.append(data["BASKET_ID"][i])
            n = n+1
            bnum.append(n)
            
    else:
        cust = []
        bas = []
        cust.append(data["CUST_CODE"][i])
        bas.append(data["BASKET_ID"][i])
        n = 1
        bnum.append(n)
        
data['BASKET_NUM'] = bnum
```

```python
# Loop to check if the product was reordered by the user 
reorder = []
prod = []
cust = []
for i in range(len(data)):
    print(len(reorder))
    
    if data["CUST_CODE"][i] in cust:
        if data["PROD_CODE"][i] in prod:
            reorder.append(1)
        else:
            prod.append(data["PROD_CODE"][i])
            reorder.append(0)
            
    else:
        cust = []
        prod = []
        cust.append(data["CUST_CODE"][i])
        prod.append(data["PROD_CODE"][i])
        reorder.append(0)

data['reorder'] = reorder
```

```python
# To calculate the max number of orders of each user
data_max = data.groupby("CUST_CODE").max()["BASKET_NUM"]
```

```python
data = pd.merge(data, data_max, on = 'CUST_CODE', how = 'left')
data.head()
```

```python
# Including only those users who have more than 2 orders.
data = data[data['BASKET_NUM_y'] > 2]
```

```python
# Train Data ---> all the orders except the last order
train = data.loc[data['BASKET_NUM_x'] != data['BASKET_NUM_y']]
train = train.reset_index(drop = True)
# As this file is ran 4 times change this name every time --> train1, train2, train3, train4
train.to_csv("train4.csv", index=False) # as this file is ran 4 times change this name every time the file is run
```

```python
# Test Data ---> Last order of the user.
test = data.loc[data['BASKET_NUM_x'] == data['BASKET_NUM_y']]
test = test.reset_index(drop = True)
# As this file is ran 4 times change this name every time --> test1, test2, test3, test4
test.to_csv("test4.csv", index=False) 
```
